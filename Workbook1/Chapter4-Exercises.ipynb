```python
def compute_pay(hours, rate):
    if hours > 40:
        pay = 40 * rate + (hours - 40) * (rate * 1.5)
    else:
        pay = hours * rate
    return pay

hours_worked = float(input("Enter Hours: "))
hourly_rate = float(input("Enter Rate: "))

pay = compute_pay(hours_worked, hourly_rate)

print("Pay:", pay)
```

CHALLENGE:
Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

Score           Grade
[ 97, 100]      1.0
[ 93,  97)        1.25
[ 89,  93)        1.5
[ 85,  89)        1.75
[ 81,  85)        2.0
[ 77,  81)        2.25
[ 74,  77)        2.5
[ 0,   74)        3.0

Run the program repeatedly to test the various different values for input.


```python
def computegrade(s):
    if s >= 97 and s < 100:
        return '1.0'
    elif s >= 93 and s < 97:
        return '1.25'
    elif s >= 89 and s < 93:
        return '1.5'
    elif s >= 85 and s < 89:
        return '1.75'
    elif s >= 81 and s < 85:
        return '2.0'
    elif s >= 77 and s < 81:
        return '2.25'
    elif s >= 74 and s < 77:
        return '2.5'
    else:
        return '3.0'

s = int(input("Enter score:"))
print(f"Grade: {computegrade(s)}")
```
