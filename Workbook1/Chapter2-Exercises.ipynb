Write a program that uses input to prompt a user for a number and then outputs the square of the number

```python

num = int(input("Enter a number: ")) #This line prompts the user to input a number by displaying the message "Enter a number: ". The input() function takes user input as a string. Then int() is used to convert the input string into an integer because we want to perform arithmetic operations on this input. The integer value entered by the user is stored in the variable num.
a = num * num #This line calculates the square of the number entered by the user. It multiplies the value of num by itself, effectively squaring it, and assigns the result to the variable a.
print("The square of", num, "is", a) #print(f"The square of {num} is {a}")

```

Write a program which prompts the user for a Celsius temperature, convert the temperature to Fahrenheit, and print out the converted temperature.

```python
#Write a program which prompts the user for a Celsius temperature, convert the temperature to Fahrenheit, and print out the converted temperature.
celcius = float(input("Enter celcius:"))#This line prompts the user to input a temperature in Celsius by displaying the message "Enter celcius:".
ftoc = (celcius * 9/5)+32 #This line calculates the Fahrenheit equivalent of the Celsius temperature entered by the user.
print("Celcius to fahrenheit is:", ftoc) #This line prints out the converted temperature.
```
