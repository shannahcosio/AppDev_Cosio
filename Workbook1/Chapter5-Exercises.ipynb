```python
a = 2
b = 2.0
c = a

# Using `is` to compare variables
print("Using 'is':")
print("a is b:", a is b)  # Should be False since they are of different types
print("a is c:", a is c)  # Should be True since they reference the same object

# Using `==` to compare values
print("\nUsing '==':")
print("a == b:", a == b)  # Should be True since the values are equal
print("a == c:", a == c)  # Should be True since the values are equal
```

CHALLENGE 1:
Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

Enter a number: 4
Enter a number: 5
Enter a number: bad data
Invalid input
Enter a number: 7
Enter a number: done
16 3 5.333333333333333

```python
total = 0
count = 0

while True:
    user_input = input("Enter a number: ")
    if user_input.lower() == "done":
        break
    try:
        number = float(user_input)
        total += number
        count += 1
    except ValueError:
        print("Invalid input")
        continue

if count > 0:
    average = total / count
    print(total, count, average)
else:
    print("No numbers were entered.")
```

CHALLENGE 2:
Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

```python
numbers = []

while True:
    user_input = input("Enter a number: ")
    if user_input.lower() == "done":
        break
    try:
        number = float(user_input)
        numbers.append(number)
    except ValueError:
        print("Invalid input")
        continue

if numbers:
    maximum = max(numbers)
    minimum = min(numbers)
    print("Maximum:", maximum)
    print("Minimum:", minimum)
else:
    print("No numbers were entered.")
```
